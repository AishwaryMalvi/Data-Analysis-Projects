```python
## How do the Average Views, Likes, and Comments are compared between different Album Types

df = data.groupby('Album_type')[['Likes', 'Views', 'Comments']].mean()

df
```

```python
type(df)
```

```python
df = df.reset_index()
```

```python
df_melted = pd.melt(df, id_vars = 'Album_type', var_name = "Attribute", value_name = 'Total' )

df_melted
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (9,4))
sns.barplot( x = 'Album_type', y = 'Total', hue = 'Attribute', data = df_melted )
plt.show() 
```

```python
## Top 5 YouTube Channels - based on the Views

c_view = data.groupby('Channel')['Views'].sum().sort_values(ascending=False)
c_view
```

```python
c_view = c_view.reset_index()
c_view
```

```python
type(c_view)
```

```python
c_view['Channel'] = c_view['Channel'].str.replace('$', r'\$', regex=False)

sns.barplot(x="Views", y="Channel", data=c_view, color='black')
plt.title('Top 5 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```
